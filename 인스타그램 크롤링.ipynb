```python
#인스타그램 댓글가져오기
#특정 태그를 한 게시물 주소 저장

import requests
import time
import pandas as pd
from bs4 import BeautifulSoup
#from selenium import webdriver
import selenium.webdriver as webdriver
from selenium.webdriver.common.keys import Keys
from operator import eq

# chromedriver 실행시간
bf_now = time.localtime()
ds = "%04d-%02d-%02d %02d:%02d:%02d" % (bf_now.tm_year, bf_now.tm_mon, bf_now.tm_mday, bf_now.tm_hour, bf_now.tm_min, bf_now.tm_sec)

# 페이지 스크롤을 위해 임시 변수 선언
pagedowns = 0

# data 담을 list
key = []
cr = []

#chrome_driver = 'C:/chromedriver'
#driver = webdriver.Chrome('C:/chromedriver')

# 인스타그램 #서울 해시테그
url = 'https://www.instagram.com/explore/tags/seoul/'
resp = requests.get(url)


# chromedriver 실행화면 숨기기
# chromedriver의 attribute에 headless 추가한다.
options = webdriver.ChromeOptions()
# options.add_argument('headless')
# options.add_argument('disable-gpu')
driver = webdriver.Chrome('C:/chromedriver', chrome_options=options)

# chromedriver 실행하기
driver.get(url)

# 2초 대기
driver.implicitly_wait(2)

# chromedriver 스크롤
elem = driver.find_element_by_tag_name("body") 

# 페이지 스크롤 대기시간
SCROLL_PAUSE_TIME = 0.5

while pagedowns < 2:
    src = driver.page_source
    soup = BeautifulSoup(src)
    # 게시물 주소
    a = soup.find_all('div',class_='Nnq7C weEfm')
    
    for j in a:
        for i in j.find_all('a'):
            
            # 리스트에 중복된 값은 제외하고 넣는다.
            if i.get('href') not in key:
                key.append(i.get('href'))
    
    elem.send_keys(Keys.PAGE_DOWN)
    
    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)
    
    pagedowns += 1

    
# 인스타그램 게시글의 해당 댓글
for k in key:
    url = 'https://www.instagram.com' + k
    resp = requests.get(url)
    driver.get(url)
    src = driver.page_source    
    soup = BeautifulSoup(src)

    # 댓글 위치
    cmt = soup.find_all('div',class_='C4VMK')
    
    for j in cmt:
        for tag in j.find_all('span'):
            coment = tag.text
            cr.append(coment)
            
# chromedriver 실행시간
aft_now = time.localtime()
de = "%04d-%02d-%02d %02d:%02d:%02d" % (aft_now.tm_year, aft_now.tm_mon, aft_now.tm_mday, aft_now.tm_hour, aft_now.tm_min, aft_now.tm_sec)
    
# 리스트 출력
# print(cr)
print(ds)
print(de)
print(len(cr))
print(pagedowns)

# csv 파일로 저장
dataframe = pd.DataFrame(cr, columns=['coment'])
dataframe.to_csv("Desktop/ins.csv", mode='w')

# 드라이버를 종료한다.
#driver.close()   
```

```python
import schedule
import time
def job():
    now = time.localtime()
    current = "%04d-%02d-%02d %02d:%02d:%02d" % (now.tm_year, now.tm_mon, now.tm_mday, now.tm_hour, now.tm_min, now.tm_sec)
    print("Current time = ", str(current))
    
schedule.every(10).seconds.do(job)

while True:
    schedule.run_pending()
    time.sleep(1)
```

```python
import schedule
import time
def job():
    now = time.localtime()
    current = "%04d-%02d-%02d %02d:%02d:%02d" % (now.tm_year, now.tm_mon, now.tm_mday, now.tm_hour, now.tm_min, now.tm_sec)
    print("Current time = ", str(current))
    
schedule.every(10).seconds.do(job)

while True:
    schedule.run_pending()
    time.sleep(1)
```
